### **3.1. What are the geographic locations of job postings across the United States?**<a id="Question1"></a>

 First as the states and cities are in one cell, we would like to extract only the cities for further processing by using split.

```python
postings = postings.withColumn("split_loc", F.split("job_location", ","))
posting_city = postings.withColumn("cities", col("split_loc").getItem(0))
post_loc = posting_city.select('cities').groupby('cities').agg(F.count('cities').alias('count'))
```

As we do not have longitude and latitude in our dataset, we imported another table which include the cities longitude and latitude so that we can plot a graph to show the geographical distribution of the job postings.

```python
# Registering DataFrame views
us_cities = spark.read.format("csv")\
  .option("header", "true")\
  .option("inferSchema", "true")\
  .option("multiline","true")\
  .load('gs://ba843-team1/notebooks/jupyter/uscities.csv')
```

We then merged the two table together and only extracting the useful information for plotting graph, also we drop the cities that is not in the main continent of the US, such as alaska and hawaii. This is because the number of job positing in these area are little and that by dropping these job postings, we can have a clearer view on how the job postings distributed in the main continent. 

```python
#!pip install contextily
post_loc = post_loc.join(us_cities, on = post_loc['cities'] == us_cities['city'], how = 'left_outer')
post_loc = post_loc.select('cities','count','lat','lng')
post_loc = post_loc.na.drop()
post_loc.createOrReplaceTempView("post_loc")
# Filtering out cities not in the main continent of the US
post_loc = spark.sql("""SELECT *
                        FROM post_loc
                        WHERE lng < 0 AND lng >= -130 AND lat < 50 AND lat > 22""")
post_loc = post_loc.toPandas()
import geopandas as gpd
import contextily as ctx
gdf = gpd.GeoDataFrame(post_loc, geometry=gpd.points_from_xy(post_loc.lng, post_loc.lat))
# Set the coordinate system to WGS84 (latitude/longitude)
gdf.set_crs(epsg=4326, inplace=True)
# Convert to Web Mercator for contextily
gdf = gdf.to_crs(epsg=3857)
# Plotting
ax = gdf.plot(figsize=(10, 10), alpha=0.01)
ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)
ax.set_axis_off()
```

**Critical Summary:** Most of the job postings mainly concentrated in the east coast and big cities, such as Washington, New York and Chicago. There are not much job posting from the central of the US to the West Coast. In the West Coast, most of the job posting are from Los Angeles, and the Bay area. 

### **3.2 What are the most in-demand Skills?**<a id="Question3"></a>

```python
# Unpack Skills arrays
skills_exploded = skills.withColumn("skill", F.explode("job_skills"))
skills_exploded.cache()

# Overview
job_count = skills.count()
skills_count = skills_exploded.count()
print(f"There are {job_count} job listings in the dataset with a total of {skills_count} skills referenced. Each job has appoximately {skills_count/job_count:.0f} skills mentioned in its listing")

# Finding top skills
skills_top_10 = skills_exploded.groupBy("skill").agg(F.count("skill").alias("count")).orderBy("count",ascending=False).limit(20).toPandas()
top_skills = list(skills_top_10.skill.unique())

# Display Top 10 Skills
skills_top_10.head(10)
```

Since there are a lot of duplicates, we will use regex to refine our match and eliminate repetitions

```python
# Using regex to match synonymous skills - Top 10 w/o duplicates

skills_exploded = skills_exploded.withColumn("Match", F.regexp_extract("skill", "\w*((?i)communication|teamwork|leadership|customer|problem|collaboration|training|detail|time|microsoft)\w*", 1))
skills_exploded = skills_exploded.withColumn("Match", F.upper(F.col("Match")))                  
skills_exploded = skills_exploded.filter(skills_exploded.Match != "")
skills_exploded.cache()                  

top_10 = skills_exploded.groupBy("Match").agg(F.count("Match").alias("Count")).toPandas()                  
                                    
# Renaming Skills
import matplotlib.ticker as mtick
df = top_10.loc[:]
df["Match"] = df["Match"].map({"COMMUNICATION":"Communication", "TEAMWORK":"Teamwork", "LEADERSHIP":"Leadership", "CUSTOMER":"Customer Service", "PROBLEM":"Problem Solving", "COLLABORATION":"Collaboration", 
                   "TRAINING":"Training/Instruction", "DETAIL":"Attention to Detail", "TIME":"Time Management", "MICROSOFT":"MS Office"})
df = df.sort_values(by="Count", ascending=True)
df["percentage"] = df["Count"]*100/job_count
# Plotting
import matplotlib.pyplot as plt
df.plot.barh(x = "Match", y="percentage", legend=None)
plt.xlabel('% of Jobs', fontsize=14)
plt.ylabel(None)
plt.gca().xaxis.set_major_formatter(mtick.PercentFormatter())
plt.xticks(fontsize=10)

plt.title('Top 10 Skills Demanded by Employers in 2024');
```

**Critical Summary:** All the top skills currently required by employers are ones that have always been important to the workplace. Fundamental skills such as communication, collaboration, and problem solving are still the most important abilities needed to excel in the professional space. With the exception of MS Office, there are no technical skills in the Top 10, indicating a perennial preference for employees with well-establish essential skills, with technical qualifications only being considered after these basic professional tools.     

### **3.3 How many skills does an applicant need?**<a id="Question4"></a>

```python
## Skill Count Distribution
import numpy as np
import matplotlib.pyplot as plt

skills = skills.withColumn("skill_count", F.size('job_skills'))
skills.cache()

# Finding middle 98% of skill_count
sc_lower  = skills.select(F.percentile_approx(F.col("skill_count"), 0.00)).collect()[0][0] # 2nd percentile
sc_upper = skills.select(F.percentile_approx(F.col("skill_count"), 0.98)).collect()[0][0] # 98th percentile

# Extracting Data
skills_np = np.array(skills.select("skill_count").collect())
skills.unpersist()
```

```python
# Plot Skill Count Histogram

def outlier_aware_hist(data, lower=False, upper=False, bins='auto', density=False):
    # Creates bounded histogram
    n, bins, patches = plt.hist(data, range=(lower, upper), bins=bins, density=density)
    plt.xlabel('# of Skills Listed', fontsize=14)
    plt.ylabel('Jobs', fontsize=18)
    plt.title('# of Skills required for Jobs in 2024', fontsize=18);

    if lower:
        n_lower_outliers = (data < lower).sum()
        patches[0].set_height(patches[0].get_height() + n_lower_outliers)
        patches[0].set_facecolor('c')
        patches[0].set_label('Lower outliers: ({:.2f}, {:.2f})'.format(data.min(), lower))

    if upper:
        n_upper_outliers = (data > upper).sum()
        patches[-1].set_height(patches[-1].get_height() + n_upper_outliers)
        patches[-1].set_facecolor('m')
        patches[-1].set_label('Upper outliers: ({:.2f}, {:.2f})'.format(upper, data.max()))

    if lower or upper:
        plt.legend()

outlier_aware_hist(skills_np, lower = sc_lower, upper=sc_upper, bins=25)
```

**Critical Summary:** Jobs require 21 skills on average, with the bottom 98% ranging from 1 to 52 skills. The distibution is right tailed, indicating employer's preference for higher a number of skills in its candidates. Let us take a closer look at the Top 20 skills demanded by employers.

We will use three different industry lists to match each job summary with an industry. We repeat thrice to maxmize our chances of a match, and coalsece to get a single industry per job posting 

```python
# Extract Industry from Summaries

# Match with Standard List of Industries
industry = industry.withColumn(
    'ICB_Industry', F.upper(F.regexp_extract('job_summary_cleaned', "\w*((?i)" + '|'.join(icb_industries) + ")\w*",1)))

# Match with Lower Level Sectors for Rows w/o Industry
industry_sector = industry.where((F.col('ICB_Industry').isNull()) | (F.col('ICB_Industry') == ""))
industry_sector = industry_sector.select('job_link', 'job_summary_cleaned')
industry_sector = industry_sector.withColumn(
    'ICB_Sector', F.upper(F.regexp_extract('job_summary_cleaned', "\w*((?i)" + '|'.join(icb_supersector) + ")\w*",1)))

# Match with lowest level sectors stems for Rows w/o Industry or Sector
industry_tokens = industry_sector.where((F.col('ICB_Sector').isNull()) | (F.col('ICB_Sector') == ""))
industry_tokens = industry_tokens.select('job_link', 'job_summary_cleaned')
industry_tokens = industry_tokens.withColumn(
    'ICB_Token', F.upper(F.regexp_extract('job_summary_cleaned', "\w*((?i)" + '|'.join(icb_tokens) + ")\w*",1)))

# Merge all 3 dataframes & coalesce
industry_merged = industry.join(industry_sector.select("job_link", "ICB_Sector"), how='left', on='job_link')
industry_merged_2 = industry_merged.join(industry_tokens.select("job_link", "ICB_Token"), how='left', on='job_link')
industry_merged_2.cache()

# Mapping sectors and tokens with Industry
my_dict = dict.fromkeys(['TECHNOLOGY', 'SOFTWAR'], 'Technology')
my_dict.update(dict.fromkeys(['TELECOMMUNICATIONS'], 'Telecommunications'))
my_dict.update(dict.fromkeys(['HEALTHCARE', 'PERSONAL CARE','HEALTH CARE','PHARMACEUT', 'CARE'], 'Healthcare'))
my_dict.update(dict.fromkeys(['FINANCE', 'FINANC', 'BANK', 'INVEST', 'BROKERAG', 'CREDIT'], 'Finance'))
my_dict.update(dict.fromkeys(['REAL ESTATE'], 'Real Estate'))
my_dict.update(dict.fromkeys(['TRAVEL AND LEISURE', 'CONSUMER PACKAGED GOODS', 'RETAIL','GROCERY' 'CONSUM', 'HOUSEHOLD'], 'Consumer'))
my_dict.update(dict.fromkeys(['BASIC MATERIALS', 'AUTOMOBILE', 'CHEMICALS', 'INDUSTRIAL', 'CHEMIC', 'METAL', 'INDUSTRI','ENGIN', 'AEROSPAC'], 'Industrials'))
my_dict.update(dict.fromkeys(['UTILITIES', 'ENERGY', 'ENERGI','COAL', 'ELECTR', 'OIL'], 'Energy'))
my_dict.update(dict.fromkeys(['MEDIA'], 'Media'))
my_dict.update(dict.fromkeys(['INSURANCE'], 'Insurance'))
my_dict.update(dict.fromkeys(['CONSTRUCTION', 'CONSTRUCT'], 'Construction'))
my_dict.update(dict.fromkeys(['DEFENS', 'GA'], 'Other'))

industry_merged_3 = industry_merged_2

# Remap names
def replace(column, value):
    return F.when(column != value, column).otherwise(F.lit(None))

for col in industry_merged_3.columns[2:]:
    industry_merged_3 = industry_merged_3.withColumn(col, replace(F.col(col), ""))
industry_final = industry_merged_3.withColumn("Industry", F.coalesce(F.col("ICB_Industry"), F.col("ICB_Sector"),F.col("ICB_Token")))

industry_final.cache()
industry_final.show()
```

### **4.4 Applying NER on Job Summaries to extract useful terms**<a id="Questions"></a>

**Named Entity Recognition**

Since our summaries table contains a lot of valuable information that we cannot extract using simple regex methods, we opted to use two pretrained NER models to extract information about tasks, tools and key words.

We initally planned on using the Spark NLP library, but found that our NLP workload would take about 20x longer when run on a CPU-only cluster. So for this phase, we used python on Colab GPUs to run the pretrained models on our dataset.

We ran two models - 
* tomaarsen/span-marker-bert-base-uncased-cross-ner - Tags tasks, fields, products and programming languages
* tomaarsen/span-marker-bert-base-uncased-keyphrase-inspec - Tags key words and phrases that are the most important given the context

```python
### NER Code

!pip install -U span_marker --quiet
import pandas as pd
import time
from span_marker import SpanMarkerModel

# Load Data
df = pd.read_csv('Data_Jobs_Cleaned.csv')
df = df.dropna(subset = ['job_summary_cleaned']).reset_index(drop=True) # drop jobs w/o summary

import torch
print(f"GPU is available: {torch.cuda.is_available()}")

# Load Models
model1 = SpanMarkerModel.from_pretrained("tomaarsen/span-marker-bert-base-uncased-cross-ner")
if torch.cuda.is_available():
  model1.cuda()
print(f'Model Device: {model1.device}')

model2 = SpanMarkerModel.from_pretrained("tomaarsen/span-marker-bert-base-uncased-keyphrase-inspec")
if torch.cuda.is_available():
  model2.cuda()
print(f'Model Device: {model2.device}')

# Function to trim JD to middle 50%, since the NER Models can only handle 512 tokens at once
def middle(desc):
  length = len(desc)
  start = round(length*0.4)
  end = round(length*0.9)
  return desc[start:end]

# Limiting text into manageable sizes
df['middle'] = df['job_summary_cleaned'].apply(middle)

# Running Models
df['entities_dicts'] = df['middle'].apply(lambda x: model1.predict(x))
df['key_words'] = df['middle'].apply(lambda x: model2.predict(x))

df["entities_dicts"] = data_jobs_ner["entities_dicts"].apply(eval) # convert these two cols fromn strings to Lists of dictionaries
df["key_words"] = data_jobs_ner["key_words"].apply(eval)

data_jobs_ner.head()
```

```python
# Extracting values from NER dictionaries

def extractor(entities_dicts, label):
  matches = []
  for i in entities_dicts:
    if label == i['label']:
      matches.append(i["span"])
  return matches

data_jobs_ner["Tools"] = data_jobs_ner["entities_dicts"].apply(lambda x:extractor(x, 'product')) 
data_jobs_ner["Field"] = data_jobs_ner["entities_dicts"].apply(lambda x:extractor(x, 'field')) 
data_jobs_ner["Task"] = data_jobs_ner["entities_dicts"].apply(lambda x:extractor(x, 'task')) 
data_jobs_ner["Lang"] = data_jobs_ner["entities_dicts"].apply(lambda x:extractor(x, 'programlang')) 
data_jobs_ner["Org"] = data_jobs_ner["entities_dicts"].apply(lambda x:extractor(x, 'organisation')) 
data_jobs_ner["key_words_list"] = data_jobs_ner["key_words"].apply(lambda x:extractor(x, 'KEY')) 

## Save file 
data_jobs_ner.to_parquet("Data_Jobs_Final.parquet")
```

```python
from pyspark.sql import Window
from pyspark.sql.functions import col, rank
#Find out the highest number of industries in the dataframe and pick the top 3 industries as analysis
# Define a window specification partitioned by 'job_cat' and ordered by 'Industry_count' descending
window_spec = Window.partitionBy('job_cat').orderBy(col('Industry_count').desc())

# Add a rank column based on the order within each partition
ranked_df = grouped_df.withColumn('rank', rank().over(window_spec))

# Filter the DataFrame to keep only the top three industries per job category
top_three_per_job_cat = ranked_df.filter(col('rank') <= 3)
top_three_per_job_cat = top_three_per_job_cat.groupBy('Industry').agg(count('Industry').alias('Industry_count'))
top_three_per_job_cat.orderBy('Industry_count', ascending=False).show(3)
```

From the table, the 2 most common industries that has hourly salaries extracted are Technology and Finance. Hence, I would use these two industries to explore the difference in hourly salaries given they are in same job categories, which expect to perform similar duties. 

```python
# Calculate the mean hourly pay for each industry within each job category
from pyspark.sql.functions import mean
from pyspark.sql.functions import col, abs, desc
mean_hourly_pay_df = final_df.groupBy('job_cat', 'Industry') \
                              .agg(mean('HourlyPay').alias('Industry_mean_salary'))
# Filter the DataFrame for 'Finance' and 'Technology' industries separately
finance_df = mean_hourly_pay_df.filter(col('Industry').contains('Finance')).orderBy('job_cat')
technology_df = mean_hourly_pay_df.filter(col('Industry').contains('Technology')).orderBy('job_cat')
# Subtract mean salary in 'Technology' from mean salary in 'Finance' for each job category
difference_df = finance_df.alias('finance').join(technology_df.alias('technology'), 'job_cat', 'inner') \
                          .withColumn('Salary_difference', abs(col('finance.Industry_mean_salary') - col('technology.Industry_mean_salary')))
                          
# Select the top 10 job categories with the highest absolute differences
top_10_differences = difference_df.orderBy(desc('Salary_difference')).limit(10)

# Extract job categories
job_categories = top_10_differences.select('job_cat').rdd.map(lambda row: row[0]).collect()

# Extract mean salaries for 'Finance' and 'Technology' for each job category
finance_salaries = top_10_differences.select('finance.Industry_mean_salary').rdd.map(lambda row: row[0]).collect()
technology_salaries = top_10_differences.select('technology.Industry_mean_salary').rdd.map(lambda row: row[0]).collect()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(len(job_categories))
width = 0.35
fig, ax = plt.subplots(figsize=(10, 6))
bar1 = ax.bar(x - width/2, finance_salaries, width, label='Finance', color='blue')
bar2 = ax.bar(x + width/2, technology_salaries, width, label='Technology', color='green')

# Set xticks to the middle of the grouped bars
ax.set_xticks(x)
ax.set_xticklabels(job_categories, rotation=90, ha='center')
ax.set_xlabel('Job Categories')
ax.set_ylabel('Mean Hourly Pay')
ax.set_title('Mean Hourly Pay by Industry and Job Category')
ax.legend()
plt.tight_layout()
plt.show()
```

Based on the graph above, it's evident that there are significant discrepancies in hourly wages between industries within the same job categories, despite their expected similar duties. In the technology sector, higher salaries are typically associated with more specialized and managerial roles. Conversely, within the finance industry, positions commanding higher wages often encompass less technical responsibilities. For instance, roles like healthcare analysts, tasked with overseeing the healthcare aspect within the firm, or senior marketing analysts focused on promoting the company, contribute to the higher wage bracket in finance.
